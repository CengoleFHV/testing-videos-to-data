```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
a_frame_chrome_data = pd.read_csv("../data/2023_06_16_Aframe50chrome_data_20230617181555.csv")
a_frame_firefox_data = pd.read_csv("../data/2023_06_16_Aframe50firefox_data_20230617181555.csv")
unity_data = pd.read_csv("../data/2023_06_16_Unity50_data_20230617181555.csv")
godot_data = pd.read_csv("../data/2023_06_16_Godot50_data_20230617181555.csv")
```

```python
a_frame_chrome_delay_data = np.array(a_frame_chrome_data["monitor"] - a_frame_chrome_data["leds"])
a_frame_firefox_delay_data = np.array(a_frame_firefox_data["monitor"] - a_frame_firefox_data["leds"])
unity_delay_data = np.array(unity_data["monitor"] - unity_data["leds"])
godot_delay_data = np.array(godot_data["monitor"] - godot_data["leds"])
```

```python
delay_data_dict = {
    "A-Frame (Chrome)": a_frame_chrome_delay_data,
    "A-Frame (Firefox)": a_frame_firefox_delay_data,
    "Unity": unity_delay_data,
    "Godot": godot_delay_data
}
```

```python
fig = plt.figure(figsize=(10,7))

x_col = "Zeit (ms)"
y_col = "Framework"
title = f"Zeitdifferenz Input Output (n={a_frame_chrome_delay_data.size})"

ax = fig.add_subplot()

ax.boxplot(delay_data_dict.values(), labels=delay_data_dict.keys(), vert=False)

ax.set_xlabel(x_col)
ax.set_ylabel(y_col)
ax.set_title(title)

ax.xaxis.set_minor_locator(AutoMinorLocator())

plt.show()
```

```python
df = pd.DataFrame(delay_data_dict)

fig = plt.figure(figsize=(8,2))
ax = plt.subplot(911, frame_on=False)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

pd.plotting.table(ax, df)

plt.show()
```

```python
stats = df.describe()

fig = plt.figure(figsize=(10,2))
ax = plt.subplot(911, frame_on=False)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

pd.plotting.table(ax, stats)

plt.show()
```
